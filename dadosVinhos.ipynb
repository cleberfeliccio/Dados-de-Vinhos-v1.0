```python
from sklearn.svm import SVC
# Aqui importamos o modelo SVC (Support Vector Classifier), que é a versão de classificação da Máquina de Vetores de Suporte.
# Ele é um dos algoritmos mais poderosos para: separação de classes,hiperplanos de decisão,dados complexos com fronteiras não lineares.

svm = SVC()
#Aqui criamos o modelo SVC com os parâmetros padrão, que são:
# kernel="rbf" (radial basis function)
C=1
# gamma="scale"
# Esse kernel RBF permite capturar relações não lineares na tabela.

svm.fit(X_train, y_train)
# Treina o modelo com:
# X_train → suas variáveis explicativas
#y_train → a classe alvo (labels)
# Durante o treino, o algoritmo busca o melhor hiperplano que separa as classes com a maior margem possível.

print(svm.score(X_test, y_test))
# O método .score() faz: prever classes de X_test/comparar com y_test/retornar a acurácia
# Ou seja, a porcentagem de acertos do SVM no conjunto de teste.
```

```python
from sklearn.ensemble import RandomForestClassifier
# Aqui importamos o algoritmo Random Forest, um dos modelos mais fortes para classificação.
# Ele funciona criando várias árvores de decisão e combinando todas para dar uma resposta mais estável e precisa.

rf = RandomForestClassifier()
# Aqui criamos o modelo com parâmetros padrão, como:
# n_estimators=100 → 100 árvores na floresta
# criterion="gini" → métrica de pureza
# max_depth=None → árvores crescem livremente
# bootstrap=True → usa amostragem com reposição
# O modelo padrão já funciona muito bem para muitos problemas.

rf.fit(X_train, y_train)
# Você treina-se o Random Forest com os dados de treino:
# X_train → características (variáveis)
# y_train → classes (labels)
# Durante o treino, o modelo:
# cria várias árvores de decisão aleatórias
# treina cada uma com uma parte dos dados
# combina os resultados das árvores para fazer previsões

print(rf.score(X_test, y_test))

# O .score() retorna a acurácia, ou seja:
#porcentagem de acertos do modelo no conjunto de teste.
# Exemplo de saída: 1,0
# Isso significa 100% de acurácia — um ótimo resultado.
```

```python
from sklearn.model_selection import cross_val_score
# Essa função permite rodar validação cruzada, uma técnica fundamental para avaliar modelos de Machine Learning.

scores = cross_val_score(LogisticRegression(), X_scaled, y, cv=5)
# Aqui usamos:
# ✔ Modelo usado
# LogisticRegression()  → Treina um modelo de regressão logística.

# ✔ Dados usados
# X_scaled → variáveis de entrada normalizadas
# y → variável alvo (target)

# ✔ Número de divisões
# cv=5 → validação cruzada com 5 folds:

# O conjunto de dados é dividido assim:
# 5 partes do dataset/4 partes para treino/1 parte para teste
# repete 5 vezes, trocando o conjunto de teste a cada rodada
# Isso evita o problema de: depender só de um único train/test split/modelo “ganhar sorte” em uma divisão específica
# Com cross-validation, você obtém uma avaliação mais estável e mais confiável.


scores.mean()
#Isso pega a média das 5 avaliações.
# Nosso modelo tem aproximadamente 98% de acurácia média
# Muito mais confiável que uma única métrica.
```
